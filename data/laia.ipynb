```python
import networkx as nx
G = nx.Graph()

# Agregar los nodos y las aristas
for _, row in df.iterrows():
    # Agregar el nodo (persona)
    node = row['id']
    G.add_node(node)
    
    # Agregar las aristas (relaciones de amistad)
    for friend in row['friend_registration']:
        G.add_edge(node, friend)


print(G.number_of_edges())
```

```python
connected_components = nx.connected_components(G)
print(type(connected_components))
# Filtrar componentes con más de 1 nodo
filtered_components = [comp for comp in connected_components if len(comp) > 1]
print(len(filtered_components))

# Imprimir las componentes filtradas
for i, component in enumerate(filtered_components, 1):
    print(f"Componente conectada {i}: {component}")
    
```

```python
import os
components_eliminar : list[str] = []
# Suponiendo que 'filtered_components' es una lista de listas o ids de componentes
# Abrir el archivo para escribir los componentes conectados
with open('grups.txt', 'w') as file:
    for i, component in enumerate(filtered_components, 1):
        file.write(f"Componente conectado {i}: {component}")
        file.write('\n')
        component_prova = component.pop()
        components_eliminar.append((component_prova))
        
print(len(df))
print(components_eliminar)
df_result = df[-df['id'].isin(components_eliminar)]
print(len(df_result))
```

```python
df["primary_challenge"] = df["interest_in_challenges"].apply(lambda x: x[0] if x else None)
```

```python
grouped_by_multiple = df.groupby(["primary_challenge", "year_of_study"])


for (challenge, year), group in grouped_by_multiple:
    print(f"Grupo: {challenge}, Año de Estudio: {year}")
    print(len(group))    
    print("\n" + "="*40 + "\n")
```

```python
grouped_by_multiple = df.groupby(["primary_challenge", "year_of_study", "experience_level"])


groups = {}

for (challenge, year, experience_level), group in grouped_by_multiple:
    # drop columns inutils
 
    projected_group = group.drop(['email', 'age', 'shirt_size', 'university', 'dietary_restrictions', 'id', 'programming_skills', 'hackathons_done', 'interests', 'preferred_role', 'objective', 'preferred_languages', 'preferred_team_size', 'availability', 'introduction', 'technical_project', 'future_excitement','fun_fact', 'friend_registration', 'interest_in_challenges'], axis=1)    
    print(f"Grupo: {challenge}, Año de Estudio: {year}, Experience: {experience_level}")
    print(len(group))    

    groups[(challenge, year, experience_level)] = [projected_group.iloc[i:i + 4] for i in range(0, len(projected_group), 4)]
    print(groups[(challenge, year, experience_level)])

    print("\n" + "="*40 + "\n")
```
